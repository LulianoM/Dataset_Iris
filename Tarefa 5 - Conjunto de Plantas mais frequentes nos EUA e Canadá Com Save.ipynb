```python
import pandas as pd
import numpy as np
```

## Importação dos dados

```python
def read_listas(name):
    with open(name, 'r') as f:
        lines = f.readlines()
    states = []
    for line in lines:
        states.append(line.strip().split(',')[1:])
    return states
```

```python
plantas = read_listas("plants.data")
```

```python
plantas[0]
```

## Convertendo a lista para uma matriz binária

```python
#!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(plantas).transform(plantas)
```

## Transformando a matriz binária em um DataFrame

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

## Encontrando conjunto de dados mais frequentes

```python
from mlxtend.frequent_patterns import apriori
```

```python
ap = apriori(df, min_support=0.15)
```

```python
ap.head(5)
```

## Utilizando regras de associação

```python
from mlxtend.frequent_patterns import association_rules
```

```python
# printing out the association rules
rules = association_rules(ap, metric="confidence", min_threshold=0.5)
print(len(rules))
```

## Visualização de regras encontradas

```python
rules.head()
```
